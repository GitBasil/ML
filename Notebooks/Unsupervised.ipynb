```python
# 10.1 K Means
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Generate synthetic data
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Apply KMeans clustering
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# Plotting the clusters and centroids
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.5, marker='X')
plt.title('K-Means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
# 10.2 Hierarchical Clustering
from scipy.cluster.hierarchy import dendrogram, linkage
from matplotlib import pyplot as plt
from sklearn.datasets import make_blobs

# Generate synthetic data
X, _ = make_blobs(n_samples=150, centers=4, cluster_std=0.5, random_state=0)

# Perform hierarchical clustering using the Ward's method
Z = linkage(X, 'ward')

# Plotting the dendrogram
plt.figure(figsize=(10, 7))
plt.title('Hierarchical Clustering Dendrogram')
dendrogram(Z, truncate_mode='lastp', p=4, leaf_rotation=45., leaf_font_size=15., show_contracted=True)
plt.xlabel('Cluster size')
plt.ylabel('Distance')

plt.show()
from sklearn.cluster import KMeans

# Applying KMeans clustering
kmeans = KMeans(n_clusters=4, random_state=0)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# Plotting the clusters and centroids for KMeans
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.5, marker='X')
plt.title('K-Means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

# Plotting the dendrogram for Hierarchical Clustering
plt.subplot(1, 2, 2)
dendrogram(Z, truncate_mode='lastp', p=12, leaf_rotation=45., leaf_font_size=15., show_contracted=True)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Cluster size')
plt.ylabel('Distance')

plt.tight_layout()
plt.show()
```

```python
# 10.3 DBSCAN
from sklearn.cluster import DBSCAN
import numpy as np

# Applying DBSCAN clustering
dbscan = DBSCAN(eps=0.5, min_samples=5)
y_dbscan = dbscan.fit_predict(X)

# Plotting the clusters and noise points
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_dbscan, s=50, cmap='viridis', label='Clustered Data')
plt.scatter(X[y_dbscan == -1, 0], X[y_dbscan == -1, 1], s=50, color='red', label='Noise')

plt.title('DBSCAN Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()

plt.show()
```

```python
# 10.4 Dimensionality Reduction
from sklearn.datasets import make_blobs
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# Generating a higher-dimensional dataset
X_hd, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, n_features=3, random_state=0)

# Applying PCA for dimensionality reduction to 2 dimensions
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_hd)

# Applying t-SNE for dimensionality reduction to 2 dimensions
tsne = TSNE(n_components=2, learning_rate='auto', init='random', random_state=0)
X_tsne = tsne.fit_transform(X_hd)

# Plotting the original 3-dimensional data (using the first two features) for simplification
plt.figure(figsize=(18, 6))

# Original data (using the first two features)
plt.subplot(1, 3, 1)
plt.scatter(X_hd[:, 0], X_hd[:, 1], c='orange', label='Original Data')
plt.title('Original 3-Dimensional Data\n(First Two Features)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

# PCA results
plt.subplot(1, 3, 2)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c='blue', label='PCA Reduced')
plt.title('After PCA Dimensionality Reduction')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')

# t-SNE results
plt.subplot(1, 3, 3)
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c='green', label='t-SNE Reduced')
plt.title('After t-SNE Dimensionality Reduction')
plt.xlabel('t-SNE Feature 1')
plt.ylabel('t-SNE Feature 2')

plt.tight_layout()
plt.show()
```

```python
# 10.5 Autoencoders
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Flatten, Reshape
import matplotlib.pyplot as plt

# Load the Fashion MNIST dataset
(x_train, _), (x_test, _) = fashion_mnist.load_data()

# Normalize the images to the range [0, 1]
x_train = x_train.astype('float32') / 255.
x_test = x_test.astype('float32') / 255.

# Flatten the images for the Dense layers
x_train_flat = x_train.reshape((x_train.shape[0], -1))
x_test_flat = x_test.reshape((x_test.shape[0], -1))

# Define the size of the encoded representations
encoding_dim = 64  # 64 floats -> compression of factor 24.5, assuming the input is 784 floats

# Define the input layer
input_img = Input(shape=(784,))

# Define the encoder layers
encoded = Dense(encoding_dim, activation='relu')(input_img)

# Define the decoder layers
decoded = Dense(784, activation='sigmoid')(encoded)

# Map the input to its reconstruction
autoencoder = Model(input_img, decoded)

# Compile the autoencoder
autoencoder.compile(optimizer='adam', loss='binary_crossentropy')

# Train the autoencoder
autoencoder.fit(x_train_flat, x_train_flat,
                epochs=50,
                batch_size=256,
                shuffle=True,
                validation_data=(x_test_flat, x_test_flat))

# Encode and decode some digits from the test set
encoded_imgs = autoencoder.predict(x_test_flat)

# Visualize the original and reconstructed images
n = 10  # Number of images to display
plt.figure(figsize=(20, 4))
for i in range(n):
    # Display original images
    ax = plt.subplot(2, n, i + 1)
    plt.imshow(x_test[i].reshape(28, 28))
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)

    # Display reconstructed images
    ax = plt.subplot(2, n, i + 1 + n)
    plt.imshow(encoded_imgs[i].reshape(28, 28))
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
plt.show()
```


Import Statements
- `import numpy as np`: Imports NumPy, a library for numerical operations in Python, used here for data manipulation.
- `import tensorflow as tf`: Imports TensorFlow, a powerful library for machine learning and deep learning.
- `from tensorflow.keras.datasets import fashion_mnist`: Imports the Fashion MNIST dataset loader from Keras. Fashion MNIST is a dataset of Zalando's article images, consisting of a training set of 60,000 examples and a test set of 10,000 examples. Each example is a 28x28 grayscale image, associated with a label from 10 classes.
- `from tensorflow.keras.models import Model`: Imports the Model class from Keras, used to group layers into an object with training and inference features.
- `from tensorflow.keras.layers import Input, Dense`: Imports specific layers from Keras used to build the autoencoder. `Input` is used to instantiate a Keras tensor, and `Dense` is a fully connected layer.
- `import matplotlib.pyplot as plt`: Imports Matplotlib's Pyplot, a plotting library used for data visualization.

 Loading and Preprocessing Data
- `(x_train, _), (x_test, _) = fashion_mnist.load_data()`: Loads the Fashion MNIST dataset. Here, labels are discarded since we're focusing on reconstructing the input images themselves, not on classification.
- `x_train = x_train.astype('float32') / 255.`: Normalizes the training images by converting their datatype to `float32` and scaling the pixel values to the range [0, 1].
- `x_test = x_test.astype('float32') / 255.`: Similar normalization is applied to the test images.

 Data Reshaping
- `x_train_flat = x_train.reshape((x_train.shape[0], -1))`: Flattens the training images into 1D arrays so they can be input into the Dense layers of the autoencoder.
- `x_test_flat = x_test.reshape((x_test.shape[0], -1))`: Similarly, flattens the test images.

 Model Architecture
- `encoding_dim = 64`: Specifies the size of the encoded representation. Here, each 28x28 image (784 pixels) will be compressed to 64 floats.
- `input_img = Input(shape=(784,))`: Defines the input layer with shape 784, corresponding to the flattened image size.
- `encoded = Dense(encoding_dim, activation='relu')(input_img)`: The encoder layer that reduces the input dimensionality to `encoding_dim`, using ReLU activation.
- `decoded = Dense(784, activation='sigmoid')(encoded)`: The decoder layer that attempts to reconstruct the original input from the encoded representation, using sigmoid activation to ensure output values are in the range [0, 1].

 Compiling and Training the Model
- `autoencoder = Model(input_img, decoded)`: Instantiates the autoencoder model, mapping the input to its reconstruction.
- `autoencoder.compile(optimizer='adam', loss='binary_crossentropy')`: Compiles the model with the Adam optimizer and binary crossentropy loss, suitable for comparing pixel values.
- `autoencoder.fit(...)`: Trains the model on the flattened training images, using them as both inputs and targets, over 50 epochs and a batch size of 256. Validation is performed on the flattened test images.

 Visualizing Reconstructions
- `encoded_imgs = autoencoder.predict(x_test_flat)`: Uses the trained autoencoder to encode and decode the test images, resulting in their reconstructed versions.
- The `plt.figure` and subsequent plotting commands visualize the original and reconstructed images side by side for comparison. The first row displays original images, and the second row displays their reconstructions.
